```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()  
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.tree import export_text
```

```python
# Load iris dataset
iris = load_iris()
iris
```

```python
iris.target
```

```python
# Iris has 4 variables so its 4d we are taking 1st 2 variables so that we can virtualize it in 2d
x = iris.data[:,:2]
y = iris.target
```

```python
x
```

```python
y
```

```python
# split the data into train and test
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train , y_train)
```

```python
#predict
y_predict_train = dtree.predict(x_train)
y_predict_test = dtree.predict(x_test)
```

```python
# accuracy
from sklearn.metrics import accuracy_score
print('training_accuracy =' , accuracy_score(y_train , y_predict_train))
print('test_accuracy =' , accuracy_score(y_test , y_predict_test))
```

```python
# overfitting model
# but purpose is to show the vizualization how it looks like
```

```python
# plot the decision tree
plt.figure(figsize = (14,12))
plot_tree(dtree , filled = True , feature_names = iris.feature_names[:2],class_names = iris.target_names)
plt.show()
```

```python
# Post Prooning method
dtree_proon = DecisionTreeClassifier(max_depth=3) # max_depth is hyper parameter
dtree_proon.fit(x_train , y_train)
```

```python
#predict
y_predict_train_proon = dtree_proon.predict(x_train)
y_predict_test_proon = dtree_proon.predict(x_test)
# accuracy
from sklearn.metrics import accuracy_score
print('training_accuracy =' , accuracy_score(y_train , y_predict_train_proon))
print('test_accuracy =' , accuracy_score(y_test , y_predict_test_proon))
```

```python
# plot the decision tree
plt.figure(figsize = (14,12))
plot_tree(dtree_proon , filled = True , feature_names = iris.feature_names[:2],class_names = iris.target_names)
plt.show()
```

```python
# plot decision boundry

plt.figure(figsize = (10,8))
x_min , x_max = x[:,0].min() - 1 , x[:,0].max() + 1
y_min , y_max = x[:,1].min() - 1 , x[:,1].max() + 1

xx,yy = np.meshgrid(np.arange(x_min , x_max , 0.01) , np.arange(y_min , y_max , 0.01))
z = dtree.predict(np.c_[xx.ravel() , yy.ravel()])
z = z.reshape(xx.shape)

plt.contourf(xx ,yy,z,alpha = 0.8)
plt.scatter(x[:,0] , x[:,1],c = y , edgecolor = 'k')
plt.xlabel('sapel length')
plt.ylabel('sapel width')
plt.show()
```
