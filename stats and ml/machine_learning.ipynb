```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Dataset/placement.csv')
df
```

```python
plt.figure(figsize = (16,5))
plt.subplot(1,2,1)
sns.distplot(df['cgpa'])

plt.subplot(1,2,2)
sns.distplot(df['placement_exam_marks'])

plt.show()
```

# cgpa - normally distributed
# placement exam - right skewed

```python
df.shape
```

```python
#Approach 1
min_value = df['cgpa'].mean() - 3*df['cgpa'].std()
max_value = df['cgpa'].mean() + 3*df['cgpa'].std()
print(min_value , '\n' , max_value)
df_new = df[(df['cgpa'] <= max_value) & (df['cgpa'] >= min_value)]
df_new.shape
```

```python
#Approach 2
df['cgpa_zscore'] = (df['cgpa'] - df['cgpa'].mean()) / df['cgpa'].std()
df
```

```python
#Trimming
df_new2 = df[(df['cgpa_zscore']<=3) & (df['cgpa_zscore'] >= -3)]
df_new2.shape
```

```python
#capping
df['cgpa'] =np.where( df['cgpa'] <= min_value , min_value , np.where(df['cgpa'] >= max_value , max_value , df['cgpa']))
```

```python
df.describe()
```

```python
plt.figure(figsize = (16,5))
plt.subplot(1,2,1)
sns.distplot(df['cgpa'])

plt.subplot(1,2,2)
sns.distplot(df['placement_exam_marks'])

plt.show()
```

```python
# after outlier treatment final step
plt.figure(figsize = (16,5))
plt.subplot(1,2,1)
sns.distplot(df['cgpa'])

plt.subplot(1,2,2)
sns.distplot(df['placement_exam_marks'])

plt.show()
```

```python
# min = q1 - 1.5*iqr
# max = q3 + 1.5*iqr
percent25 = df['placement_exam_marks'].quantile(0.25)
percent75 = df['placement_exam_marks'].quantile(0.75)
iqr = percent75 - percent25
df.shape
```

```python
min_val = percent25 - 1.5*iqr
max_val = percent75 + 1.5*iqr
print(min_val , max_val)
```

```python
#trimming
df1 = df[(df['placement_exam_marks'] >= min_val) & (df['placement_exam_marks'] <= max_val) ]
df1.shape
```

```python
#capping
df['placement_exam_marks'] = np.where(df['placement_exam_marks'] <= min_val , min_val , np.where(df['placement_exam_marks'] >= max_val , max_val, df['placement_exam_marks']))
```
